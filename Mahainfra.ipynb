```python
import pandas as pd
```

```python
# read the input file
df = pd.read_csv("/content/SampleData.csv")
```

```python
df
```

```python
# Calculate no of leads per file
no_leads = len(df) // 3
```

```python
#Splitting file
file_1 = df.iloc[:,:no_leads]
file_2 = df.iloc[:,no_leads:2*no_leads]
file_3 = df.iloc[:,2*no_leads:]
```

```python
# save into seperate files
file_1.to_excel("output_part1.xlsx", index=False)
file_2.to_excel("output_part2.xlsx", index=False)
file_3.to_excel("output_part3.xlsx", index=False)
```
